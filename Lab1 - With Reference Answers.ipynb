## Part 0. Virtual Environment Setup

A virtual environment is an isolated Python environment that allows you to install packages without affecting your system Python installation. This is a best practice for Python development.

### Why Use Virtual Environments?
- **Isolation**: Each project can have its own dependencies
- **Version Control**: Different projects can use different versions of the same package
- **Reproducibility**: Makes it easier to share your project with others
- **Clean System**: Keeps your system Python installation clean

### Step 1: Create a Virtual Environment


## Part 1. Simple LLM

Now that we have our environment set up, let's explore how a simple Language Model works by building a basic chatbot.

````python
# import necessary libraries
import random
````

````python
class SimpleLLM:
    def __init__(self):
        # A tiny "knowledge base" of probable next responses
        # In a real LLM, this is billions of parameters, not a dictionary.
        self.probable_responses = {
            "hello": ["Hi there!", "Hello! How can I help?", "Greetings."],
            "how are you": ["I'm just code, but I'm functioning well.", "Doing great!", "I am ready to help."],
            "bye": ["Goodbye!", "See you later.", "Have a nice day."]
            # ... more patterns could be added here
        }
        self.default_responses = ["I'm not sure how to continue that sentence.", "Could you rephrase that?", "Interesting..."]

    def predict_next_sequence(self, input_text):
        """
        Simulates the 'next word prediction' logic based on the last user input.
        """
        # Clean the input to find a keyword (simulating 'understanding' context)
        clean_input = input_text.strip().lower()
        
        # Check if we have a probable continuation for this input
        possible_continuations = self.default_responses
        
        for key in self.probable_responses:
            if key in clean_input:
                possible_continuations = self.probable_responses[key]
                break
        
        # The "Prediction" step: Pick the most statistically likely (random here) completion
        return random.choice(possible_continuations)

def run_chatbot():
    # 1. Initialize the model
    llm = SimpleLLM()
    
    # 2. Define the "Context Window" (The script history)
    conversation_history = "The following is a conversation with an AI.\n"
    
    print("--- Simple LLM Chatbot (Type 'exit' to stop) ---")
    
    while True:
        user_input = input("User: ")
        if user_input.lower() == 'exit':
            break
            
        # 3. Append user input to the history
        conversation_history += f"User: {user_input}\n"
        conversation_history += "AI: " # We leave this open-ended
        
        # 4. The LLM "predicts" what comes after "AI: "
        prediction = llm.predict_next_sequence(user_input)
        
        # 5. Display the result and add it to history
        print(f"AI: {prediction}")
        conversation_history += f"{prediction}\n"

        # (Optional) Print what the LLM actually "sees" internally
        # print(f"\n[DEBUG - Internal Script State]:\n{conversation_history}\n")
````

````python
run_chatbot()
````

## Part 2. Local LLM with Ollama

In this section, we'll learn how to run a real LLM locally on your machine using **Ollama**. Ollama makes it easy to run open-source large language models locally without requiring cloud services.

### What is Ollama?

Ollama is a tool that allows you to download and run large language models on your local machine. It provides a simple API to interact with models like Llama, Mistral, and others.

### Step 1: Install Ollama

**On macOS:**
```bash

```

Or download from: https://ollama.ai/download

**On Linux:**
```bash
curl -fsSL https://ollama.ai/install.sh | sh
```

**On Windows:**
Download the installer from: https://ollama.ai/download

### Step 2: Start Ollama Service

After installation, start the Ollama service:


This starts the Ollama server in the background. You can verify it's running by opening <b> a new terminal </b> and running:

## Part 3. ChatGPT API Integration via HKBU GenAI Platform

In this section, we'll learn how to interact with powerful cloud-based LLMs like GPT-4 through API calls. This is different from Part 2 (local models) because:

- **Cloud-based**: The model runs on remote servers, not your machine
- **More powerful**: Access to state-of-the-art models like GPT-4
- **API-based**: We send requests over the internet and receive responses
- **Cost consideration**: Usually requires payment or quota management

### Why Use APIs?
- Access to the most advanced models (GPT-4, Claude, etc.)
- No local hardware requirements
- Handles updates and improvements automatically
- Scalable for production applications

### Step 1: Get Your API Key

HKBU provides access to various AI models through their GenAI platform. Here's how to get your API key:

**Steps:**
1. Go to https://genai.hkbu.edu.hk/ and log in with your HKBU credentials
2. Choose "API" from the menu
3. Click the "Generate" button and copy the key
4. **⚠️ IMPORTANT**: Keep your API key secret! Never share it publicly or commit it to GitHub

**Security Best Practice:**
In production, you should store API keys in environment variables or secure configuration files, not directly in your code.

````python
# Import the requests library for making HTTP calls
# This is the standard Python library for API interactions
import requests
````

````python
# Configuration for HKBU GenAI API
api_key =  'YOUR_API_KEY'  # Your personal API key - keep this secret!
base_url = "https://genai.hkbu.edu.hk/api/v0/rest"  # HKBU's GenAI endpoint
model_name = "gpt-4.1"  # The specific GPT model we want to use
api_version = "2024-12-01-preview"  # API version for compatibility
````

### Request and Response of ChatGPT API

````python
# 1. Prepare the headers for authentication and content type
headers = {
    "accept": "application/json",  # We expect JSON response
    "Content-Type": "application/json",  # We're sending JSON data
    "api-key": api_key,  # Authentication using our API key
}

# 2. Create the message list (conversation context)
# Each message has a 'role' (user/assistant/system) and 'content'
messages = [
    {"role": "user", "content": "Hi, how are you?"}
]

# 3. Construct the payload with messages and parameters
payload = {
    "messages": messages,
    "temperature": 0.7,  # Moderate creativity
    "max_tokens": 150,  # Limit response length
    "top_p": 1,  # Use all probable tokens
    "stream": False  # Get complete response at once
}

# 4. Build the complete URL with model deployment and API version
url = f"{base_url}/deployments/{model_name}/chat/completions?api-version={api_version}"

# 5. Send the POST request and get the response
response = requests.post(url, json=payload, headers=headers)

# Check if the request was successful
print(f"Status Code: {response.status_code}")
print(f"Response received: {'✓' if response.status_code == 200 else '✗'}")
````

````python
response.json()
````

### Step 3: Understanding the Response Structure

The API response contains several important fields:

```json
{
  "id": "unique-request-id",
  "object": "chat.completion",
  "created": timestamp,
  "model": "gpt-4.1",
  "choices": [
    {
      "index": 0,
      "message": {
        "role": "assistant",
        "content": "The actual response text"
      },
      "finish_reason": "stop"
    }
  ]}
```

**Key Fields:**
- `choices[0].message.content`: The actual AI response
- `finish_reason`: Why the response ended ("stop" = natural end, "length" = hit token limit)

### Wrap it into a function form

````python
def submit(user_message: str, system_message=None):
    if system_message is not None:
        messages = [
            {"role": "system", "content": system_message}
        ]
    else:
        messages = []
    
    messages += [
        {"role": "user", "content": user_message},
    ]

    url = f"{base_url}/deployments/{model_name}/chat/completions?api-version={api_version}"
    headers = {
        "accept": "application/json",
        "Content-Type": "application/json",
        "api-key": api_key,
    }
    payload = {"messages": messages, "temperature": 0.7, "max_tokens": 150, "top_p": 1, "stream": False}
    response = requests.post(url, json=payload, headers=headers)

    if response.status_code == 200:
        return response.json()['choices'][0]['message']['content']
    else:
        return {"error": response.status_code, "message": response.text}
````

````python
# Test the function with a simple question
print(submit('Hi, how are you?'))
````
